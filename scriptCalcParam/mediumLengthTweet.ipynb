users = rethinkDBservice.getUsersCursors()

```python
def get_users_mediumLengthTweets(tweets):
    lSum = 0
    for t in tweets:
        l = len(t['text'])
        lSum = lSum + l
    nbTweets = len(tweets)
    mediumLength = round(lSum / nbTweets)
    return mediumLength
```

for user in users:<br>
    tweets = list(rethinkDBservice.getTweetsByUserIdCursors(user["id"]))<br>
    result = get_users_mediumLengthTweets(tweets)<br>
    rethinkDBservice.updateUser(user["id"], result)
